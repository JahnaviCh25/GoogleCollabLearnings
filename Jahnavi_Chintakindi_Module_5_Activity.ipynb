I have choosen Timeseries plot with error bands as my graph chart for describiing data visualization

```python
import seaborn as sns
sns.set_theme(style="darkgrid",palette='deep')

# Load an example dataset with long-form data
fmri = sns.load_dataset("planets")

# Plot the responses for different events and regions
sns.lineplot(x="year", y="distance",
             data=fmri)
```

From the dataset planets, the Timeseries plot with error bands shows how much distance each planet travelled in that particular year based on the radial velocity and timimg variations as mentioned in the method legend. The X-axis attribute is the Year, Y-axis attribute is the distance.

Based on this graph, we can illustrate that the longest distance travelled was between the year 2005 to 2010. There was a sudden strike in the distance in 2002 to 2005 whereas from 1990 till 2000 it was flat with distance travelled less than 200. There seems to be a moderate shift in the distance travelled from 2010 till 2015.
