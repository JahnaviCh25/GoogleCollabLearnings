Data Set: Heart Failure Prediction
Summary:
This database named Heart Failure Prediction is a collection of data about the events caused by cardio vascular disease. This dataset consists of 918 observations. The attributes in the dataset include:

1. Age: age of the patient
2. Sex: sex of the patient
3. ChestPainType: chest pain type
4. RestingBP: resting blood pressure
5. Cholesterol: serum cholesterol
6. FastingBS: fasting blood sugar
7. RestingECG: resting electrocardiogram results
8. MaxHR: maximum heart rate achieved
9. ExerciseAngina: exercise-induced angina
10. Oldpeak: oldpeak = ST
11. ST_Slope: the slope of the peak exercise ST segment
12. HeartDisease: output class

```python
# Heart Disease Risk Factor - Age

import urllib.request
#Have taken the Age data from Age column converted to text file from heart.csv file
target_url = 'https://raw.githubusercontent.com/JahnaviCh25/Datasets/main/HeartperAge'
Age = []
for line in urllib.request.urlopen(target_url):
    Age.append(line.decode('utf-8'))

Age_data = [int(x.rstrip()) for x in Age] # remove line breaks
print(Age_data)
print(len(Age_data))
```

```python
# mean - add everything up and divide by n
sum(Age_data)/len(Age_data)
```

```python
# median - sort everthing and then find the middle value

Age_data.sort()
print(Age_data)
print(Age_data[0])
print(len(Age_data)/2)
print(Age_data[int(len(Age_data)/2)])
```

```python
# mode - find the most frequent occurence
most_frequent = {}
for i in Age_data:
  if i in most_frequent:
    most_frequent[i] += 1
  else:
    most_frequent[i] = 1
max(most_frequent, key=most_frequent.get)
```

```python
# plotting mean, median, mode
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

xmean = np.mean(Age_data)
xmedian = np.median(Age_data)
xmode = stats.mode(Age_data, keepdims=True)[0][0]

fig = plt.figure(figsize=(10,5))
sns.histplot(Age_data, kde=True, color='black')
plt.axvline(x=xmean, color='yellow', label='Mean', linewidth=7)
plt.axvline(x=xmedian, color='orange', label='Median', linewidth=5)
plt.axvline(x=xmode, color='pink', label='Mode')
plt.legend()
plt.show()
```

```python
# visualize outliers before and after
from random import random
from random import randint
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=100, n_features=1, noise=50, random_state=42)
print('mean before outliers:', np.mean(X))
print('var before outliers:', np.var(X))
fig, ([ax1, ax2], [ax3, ax4]) = plt.subplots(2, 2, figsize=(10, 10))
ax1.scatter(X, y)
ax2.hist(X)
ax3.boxplot(X)
sns.violinplot(ax=ax4, data=X);
```

```python
# create outliers
for i in range(30):
    factor = randint(-3, 3)
    if random() > 0.5:
        X[i] += factor * X.std()
    else:
        X[i] -= factor * X.std()

fig, ([ax1, ax2], [ax3, ax4]) = plt.subplots(2, 2, figsize=(10, 10))
ax1.scatter(X, y)
ax2.hist(X)
ax3.boxplot(X)
sns.violinplot(ax=ax4, data=X)
print('mean after outliers:', np.mean(X))
print('var after outliers:', np.var(X))
```

```python
# Skewed data
import numpy as np
from scipy.stats import skewnorm
from scipy import stats
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, sharey=True, figsize=(14, 3))
skew = 10
n = 100000
# right skew

r = skewnorm.rvs(skew, loc=0, scale=10, size=n)
ax[0].hist(r, bins=50, density=True, alpha=0.4)
ax[0].axvline(x=np.median(r), color='yellow', label='Median')
ax[0].axvline(x=np.mean(r).round(2), color='purple', label='Mean')
ax[0].legend()
print(f'right skew data mean: {np.mean(r).round(2)}, median: {np.median(r).round(2)}')

# left skew
l = skewnorm.rvs(-skew, loc=0, scale=10, size=n)
ax[1].hist(l, bins=50, density=True, alpha=0.4);
ax[1].axvline(x=np.mean(l).round(2), color='blue', label='Mean')
ax[1].axvline(x=np.median(l), color='black', label='Median')
ax[1].legend()
print(f'left skew data mean: {np.mean(l).round(2)}, median: {np.median(l).round(2)}')
```
